```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Generate some sample data
np.random.seed(42)
normal_data = np.random.normal(loc=0, scale=0.1, size=(1000, 2))
anomaly_data = np.random.uniform(low=-1, high=1, size=(50, 2))

# Combine normal and anomaly data
data = np.vstack([normal_data, anomaly_data])

# Fit Isolation Forest model
model = IsolationForest(contamination=0.05)  # 5% contamination (adjust as needed)
model.fit(data)

# Predict anomalies
anomaly_pred = model.predict(data)

# Plot data points and anomalies
plt.scatter(data[:, 0], data[:, 1], c=anomaly_pred, cmap='viridis')
plt.colorbar(label='Anomaly Score')
plt.title('Anomaly Detection with Isolation Forest')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
